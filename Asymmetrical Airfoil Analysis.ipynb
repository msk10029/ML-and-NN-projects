```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
original_data=pd.read_csv("Cl_Alpha Un.csv")
original_data.head()
```

```python
cor_data=original_data.dropna()
```

```python
cor_data.tail()
```

```python
cor_data.isna().sum()
```

```python
cor_data.info()
```

```python
data=pd.DataFrame(np.zeros((774,3),dtype = None))
data.rename(columns={0:'Reynold No', 1:'Angle of Attack', 2:'Cl'},inplace=True)
data.tail()
```

```python
r=50000
c=0
j=0
check=0
while r<400000:
    for i in range(0,43):
        data.iloc[j,1]=cor_data.iloc[i,c]
        data.iloc[j,2]=cor_data.iloc[i,c+1]
        data.iloc[j,0]=r
        j=j+1
    r=r+20000
    c=c+3
    check+=1
```

```python
data.tail()
```

```python
result=data.iloc[:,2]
result.head()
```

```python
train_set=data.copy(deep=True)
train_set.drop('Cl',axis=1,inplace=True)
train_set.head()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(train_set,result)
```
